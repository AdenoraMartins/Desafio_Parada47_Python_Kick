1- Importando dados de uma planilha excel

```python
import pandas as pd

tabela = pd.read_excel("tabela_alunos.xlsx")
display(tabela)
```

2 - Obter nome de alunos presentes

```python
por_aluno = tabela[["Nome"]].groupby("Nome").sum()
display(por_aluno)
```

3 - Definir quais alunos sairam e entraram novamente na aula

```python
por_saida = tabela[["Nome", "Atividade"]].groupby("Nome").sum()
display(por_saida)
```

4- Identificar o horário que cada aluno entrou.

```python
por_tempo = tabela[["Nome", "Hora"]].groupby(["Nome", "Hora"]).sum()
display(por_tempo)
```
